```python
# Importing all necessary libraries 
import cv2 
import os 

# Read the video from specified path 
cam = cv2.VideoCapture(r"C:\Users\patri\Dropbox\FHNW\Geomatik_Privat\_MSc\VP1\03_Rohdaten\_Aufnahmen\IGEO\Befliegungen_Nadir_BS-31-03-21\D1\DJI_0002.mp4") 

# frame 
currentframe = 2100
count = 0
endcounter = 100

while(True): 
	
	# reading from frame 
	ret,frame = cam.read() 

	if ret: 
		if count == endcounter:
            
			# if video is still left continue creating images 
			name = './data/video/20210408-Nadir/frame/frame' + str(currentframe) + '.jpg'
			print ('Creating...' + name) 

            # writing the extracted images 
			cv2.imwrite(name, frame) 

            # increasing counter so that it will 
            # show how many frames are created 
			currentframe += 1
			count =0
		else:
			count += 1            
	else: 
		break

# Release all space and windows once done 
cam.release() 
cv2.destroyAllWindows() 
```
